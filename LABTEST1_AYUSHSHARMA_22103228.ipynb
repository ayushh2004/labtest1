```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
traffic_signal=100+np.random.randn(1440)+np.random.normal(0,10,1440)
print(traffic_signal)
```

```python
smooth_traffic_signal_data=signal.savgol_filter(traffic_signal,51,3)
print(smooth_traffic_signal_data)
```

```python
plt.figure()
plt.plot(traffic_signal,label="traffic signal with noise",color="green")
plt.plot(smooth_traffic_signal_data,label="smooth traffic signal",color="blue")
plt.legend()
plt.plot()
```
